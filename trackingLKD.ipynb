```python
import sys
import matplotlib.pyplot as plt
import cv2
print cv2.__version__
scale = 1
delta = 0
ddepth = cv2.CV_16S
# needed for orb because of a bug
#cv2.ocl.setUseOpenCL(False)
import imp

import lktrack
#imp.reload(lktrack)
import glob
import numpy as np
```

```python
%matplotlib inline
img_array = []

direc = './data/'
for i in range(108):
    #fileName = direc + 'image00' + str(i+1) +'.jpg'
    fileName = direc + 'img' + str(i+1) +'.jpg'
    img_array.append(fileName)
    
#print img_array[0]
print "number of images ", len(img_array)
imgA = img_array[40:80] # images of interst showing motion

image = cv2.imread(imgA[0])
plt.imshow(image)
```

```python
# start tracker and detect interest points
lkt = lktrack.LKTracker(imgA)
lkt.detect_points()
```

```python


for i in range(len(imgA)-1):
    lkt.track_points()
    lkt.write()
#cv2.destroyAllWindows()
```

```python

direc = './trackedImages/'
imageT_array =[]
for i in range(40):
    fileName = direc + 'imgT' + str(i+1) +'.jpg'
    imageT_array.append(fileName)

for r in range(6):
    image = cv2.imread(imageT_array[5*(r+1)])
    plt.figure()
    plt.imshow(image)
```

```python
lkt = lktrack.LKTracker(imgA)

c = 0
for im,ft in lkt.track():
    print 'tracking %d features' % len(ft)
    
    if c%5==0:
        plt.figure()
        plt.imshow(im)
        #for p in ft:
        #    plt.plot(p[0],p[1],'bo')
        for t in lkt.tracks:
            plt.plot([p[0] for p in t],[p[1] for p in t])
        plt.axis('off')
        plt.show()
        
    c += 1
    
```
